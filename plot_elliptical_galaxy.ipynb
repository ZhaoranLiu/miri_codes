```python
##make Artificial galaxy###


import numpy as np
import matplotlib.pyplot as plt

# Define the size and shape of the galaxy
n = 200
a = 100
b = 50
theta = np.pi/4

# Create a meshgrid of coordinates
x, y = np.meshgrid(np.arange(n), np.arange(n))

# Define the elliptical galaxy
galaxy = ((x - n/2)*np.cos(theta) + (y - n/2)*np.sin(theta))**2 / a**2 + \
         ((x - n/2)*np.sin(theta) - (y - n/2)*np.cos(theta))**2 / b**2
galaxy[galaxy > 1] = 1

# Add some random noise
noise = np.random.normal(0, 0.05, (n, n))
galaxy += noise

# Adjust the contrast and brightness
galaxy = (galaxy - galaxy.min()) / (galaxy.max() - galaxy.min())
galaxy = galaxy * 255

# Plot the galaxy
plt.imshow(galaxy, cmap='gray')
plt.axis('off')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Define the size and shape of the galaxy
n = 200
a = 100
b = 50
theta = np.pi/4

# Create a meshgrid of coordinates
x, y = np.meshgrid(np.arange(n), np.arange(n))

# Define the elliptical galaxy
galaxy = ((x - n/2)*np.cos(theta) + (y - n/2)*np.sin(theta))**2 / a**2 + \
         ((x - n/2)*np.sin(theta) - (y - n/2)*np.cos(theta))**2 / b**2
galaxy[galaxy > 1] = 1

# Add some random noise
noise = np.random.normal(0, 0.05, (n, n))
galaxy += noise

# Adjust the contrast and brightness
galaxy = (galaxy - galaxy.min()) / (galaxy.max() - galaxy.min())
galaxy = galaxy * 255

# Plot the galaxy and ellipses
fig, ax = plt.subplots(figsize=(8,8))
ax.imshow(galaxy, cmap='gray')
ax.axis('off')

# Define the ellipse parameters
ell_a = a / 5
ell_b = b / 5
ell_theta = theta * 180 / np.pi

# Plot the ellipses
for i in range(1, 10):
    ell = Ellipse((n/2, n/2), i*ell_a, i*ell_b, ell_theta, fill=False, edgecolor='red')
    ax.add_patch(ell)
    
plt.show()
```
